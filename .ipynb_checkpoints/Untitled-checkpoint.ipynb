```python
#video detection of single color
import cv2
import numpy as np
cap = cv2.VideoCapture(0)
#set blue boundaries so that only blue objects are detected
lower_blue=np.array([78,43,46])
upper_blue=np.array([110,255,255])
while(1):  
    ret, frame = cap.read()   
    cv2.imshow('Capture', frame)  
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV) 
    #mask covers non-blue objects with black
    mask = cv2.inRange(hsv, lower_blue, upper_blue)
    cv2.imshow('Mask', mask)
    #res show only the unmasked part with color
    res = cv2.bitwise_and(frame, frame, mask=mask)
    #if res != mask:
    #    font = cv2.FONT_HERSHEY_SIMPLEX
    #    cv2.putText(res,'Blue',(10,500), font, 4,(255,255,255),2,cv2.LINE_AA)
    cv2.imshow('Result', res)   
    if cv2.waitKey(1) & 0xFF == ord('q'):      
        break
cap.release()
cv2.waitKey(1)
cv2.destroyAllWindows()
for i in range (1,5):
    cv2.waitKey(1)
```

```python
#image detection of single color
import cv2
import numpy as np
image = cv2.imread('greenchart.png')

boundaries = [
    ([35,43,46], [77,255,255]) #green
]

while(1):
    for (lower, upper) in boundaries:
        lower = np.array(lower, dtype = "uint8")
        upper = np.array(upper, dtype = "uint8")

        #mask
        hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV) 
        mask = cv2.inRange(hsv, lower, upper)
        output = cv2.bitwise_and(image, image, mask = mask)
        #if output != mask:
        #    font = cv2.FONT_HERSHEY_SIMPLEX
        #    cv2.putText(output,'Blue',(10,500), font, 4,(255,255,255),2,cv2.LINE_AA)
        cv2.imshow("images", np.vstack([image, output]))
    if cv2.waitKey(1) & 0xFF == ord('q'):      
        break

cv2.waitKey(1)
cv2.destroyAllWindows()
for i in range (1,5):
    cv2.waitKey(1)
    
```

```python
#click event for color detection
import sys
import cv2
import numpy as np
import tkinter as tk
from PIL import ImageTk, Image
imgCV = cv2.imread('greenchart.png')
hsv = cv2.cvtColor(imgCV, cv2.COLOR_BGR2HSV)

root = tk.Tk()
geometry = "%dx%d+0+0"%(imgCV.shape[0], imgCV.shape[1])
root.geometry()

#left click event
def leftclick(event):
    print("position")
    print (event.x, event.y)
    print("BGR color")
    print (imgCV[event.y, event.x])
    print("HSV color")
    print (hsv[event.y, event.x])
    greenDetection([event.y, event.x])
    
def greenDetection(location):
    boundaries = [
        #('green' ,[35,43,46], [77,255,255]), #all-green
        ('apple green', [141-20, 182-20, 0], [141+20, 182+20, 0]),
        ('aquamarine', [0, 205, 191], [102, 255, 170]),
        ('asparagus', [135-20, 169-20, 107-20], [135+20, 169+20, 107+20]),
        ('avacado', [86+20, 130+20, 3], [86-20, 130-20, 0]),
        ('bright green', [102-20, 255-20, 0], [102+20, 255, 0]),
        
    ]
    if (imgCV[location[0], location[1]][1] > imgCV[location[0], location[1]][0] and imgCV[location[0], location[1]][1] > imgCV[location[0], location[1]][2]):
        print('green')
        for (name, lower, upper) in boundaries:
            lower = np.array(lower, dtype = "uint8")
            upper = np.array(upper, dtype = "uint8")
        
            if cv2.inRange(imgCV[location[0], location[1]], lower, upper).all():
                print(name)


# import image
img = ImageTk.PhotoImage(Image.open('greenchart.png'))
panel = tk.Label(root, image = img)
panel.bind("<Button-1>", leftclick)
panel.pack(fill = "both", expand = 1)
root.mainloop()
```
